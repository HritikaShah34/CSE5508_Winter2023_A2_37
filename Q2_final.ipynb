Preprocessing

```python
def remove_stopwords(text):
	stop_words = set(stopwords.words("english"))
	word_tokens = word_tokenize(text)
	filtered_text = [word for word in word_tokens if word not in stop_words]
	return filtered_text

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


for i1 in range(len(df)):
    # print(len(df.loc[i1,'Text']))
    data = df.loc[i1,'Text']
    data = [data.lower()]

    # punc = '''!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~\n\t\\\€«§8»ş'''   
    # for ch in data:
    #   if ch in punc:
    #     data = data.replace(ch," ")
    translator = str.maketrans('', '', string.punctuation)
    data=[i.translate(translator) for i in data]

    data1=[]
    for i in data:
        data1.append(remove_stopwords(i))
        
    data2=[]

    for i in data1:
        s=""
        for j in i:
            s+=(j + ' ')
        data2.append(s)
            
    data = [" ".join(i.split()) for i in data2]
    # print(data[0])
    
    df.at[i1,'Text'] = data[0]
    
    print(len(df.loc[i1,'Text']))
    text = data[0]
    words = word_tokenize(text)

    # stemmer = PorterStemmer()

    stemmed_words = [lemmatizer.lemmatize(word) for word in words]

    # print("Stemmed words: ", stemmed_words)
    data3=[]

    s=""
    for j in stemmed_words:
        s+=(j + ' ')
        data3.append(s)
    
    df.at[i1,'Text'] = s
    

    
# print(df)
```

TF-IDF based model traing and testing

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


data = df['Text']

label = df['Category']

# sample training data
X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=0.3, random_state=42)

# initialize the classifier with TF-IDF weighting scheme
classifier = make_pipeline(TfidfVectorizer(), MultinomialNB())

# fit the model with the training data
classifier.fit(X_train, y_train)

# predict the labels of the test data
predicted_labels = classifier.predict(X_test)

# calculate the accuracy
accuracy = accuracy_score(y_test,predicted_labels)

# print the accuracy
print(f"Accuracy: {accuracy}")

```
